### Smart Energy Consumption Analyzer

```python
# 1. Stores energy consumption using appropriate variables and data types
# Let's represent each building's data as a dictionary in a list

buildings_data = [
    {"name": "Building A", "consumption": 2500},
    {"name": "Building B", "consumption": 1500},
    {"name": "Building C", "consumption": 4000},
    {"name": "Building D", "consumption": 800},
    {"name": "Building E", "consumption": 3000},
    {"name": "Building F", "consumption": 5500}
]

critical_grid_limit = 10000  # kWh

classified_buildings = []
total_consumption_so_far = 0
```

```python
# 2. Uses a loop to process energy data for multiple buildings
# 3. Uses if–elif–else to classify buildings
# 5. Uses break if total consumption exceeds a critical grid limit

print("--- Processing Energy Data ---")
for building in buildings_data:
    building_name = building["name"]
    consumption = building["consumption"]

    # Classify building based on consumption
    if consumption < 1000:
        classification = "Energy Efficient"
    elif 1000 <= consumption <= 3000:
        classification = "Moderate Consumption"
    else:
        classification = "Energy Intensive"

    classified_buildings.append({
        "name": building_name,
        "consumption": consumption,
        "classification": classification
    })

    total_consumption_so_far += consumption

    print(f"  {building_name}: {consumption} kWh - {classification}")

    if total_consumption_so_far > critical_grid_limit:
        print(f"  [WARNING] Critical grid limit ({critical_grid_limit} kWh) exceeded! Stopping processing.")
        break

print(f"Total consumption processed so far: {total_consumption_so_far} kWh")
print("------------------------------")
```

```python
# 4. Uses a function to compute total and average consumption
def calculate_consumption_stats(buildings_list):
    total_cons = sum(b["consumption"] for b in buildings_list)
    average_cons = total_cons / len(buildings_list) if buildings_list else 0
    return total_cons, average_cons


final_total_consumption, final_average_consumption = calculate_consumption_stats(classified_buildings)
```

```python
# 6. Prints a sustainability report

print("\n--- Sustainability Report ---")
print(f"Critical Grid Limit: {critical_grid_limit} kWh\n")

print("Building-wise Classification:")
for building in classified_buildings:
    print(f"  - {building['name']}: {building['consumption']} kWh ({building['classification']})")

print(f"\nOverall Total Consumption: {final_total_consumption} kWh")
print(f"Overall Average Consumption per Building: {final_average_consumption:.2f} kWh")

if total_consumption_so_far > critical_grid_limit:
    print("\nAction Required: The critical grid limit was exceeded. Immediate measures for energy reduction or grid management are recommended.")
else:
    print("\nStatus: Overall energy consumption is within the critical grid limit.")

print("-------------------------------")
```

## Define EnergySource Base Class

### Subtask:
Create the `EnergySource` base class with a constructor to initialize common attributes like source name and a method for power output, demonstrating encapsulation.


**Reasoning**:
The subtask requires defining an `EnergySource` base class with a constructor to initialize protected attributes for source name and current output, and a public method to retrieve the current output. This code block directly implements these requirements.



```python
class EnergySource:
  def __init__(self, name, current_output):
    self._name = name  # Protected attribute for source name
    self._current_output = current_output  # Protected attribute for current power output

  def get_current_output(self):
    """Returns the current power output of the energy source."""
    return self._current_output

# Example usage:
solar_panel = EnergySource("Solar Panel Array", 1500) # 1500 kW
wind_turbine = EnergySource("Wind Turbine Farm", 3000) # 3000 kW

print(f"{solar_panel._name} current output: {solar_panel.get_current_output()} kW")
print(f"{wind_turbine._name} current output: {wind_turbine.get_current_output()} kW")
```

## Derive Subclasses (Solar, Wind, Hydro)

### Subtask:
Implement `SolarEnergy`, `WindEnergy`, and `HydroEnergy` subclasses, inheriting from `EnergySource`. Each subclass will have its own constructor and potentially unique attributes. This will demonstrate inheritance.


**Reasoning**:
The subtask requires defining three subclasses (`SolarEnergy`, `WindEnergy`, `HydroEnergy`) that inherit from `EnergySource`, each with its own constructor and specific attributes, and then demonstrating their instantiation. This code block directly implements these requirements.



```python
class SolarEnergy(EnergySource):
  def __init__(self, name, current_output, panel_efficiency):
    super().__init__(name, current_output)
    self._panel_efficiency = panel_efficiency # Unique attribute for SolarEnergy

class WindEnergy(EnergySource):
  def __init__(self, name, current_output, wind_speed):
    super().__init__(name, current_output)
    self._wind_speed = wind_speed # Unique attribute for WindEnergy

class HydroEnergy(EnergySource):
  def __init__(self, name, current_output, water_flow_rate):
    super().__init__(name, current_output)
    self._water_flow_rate = water_flow_rate # Unique attribute for HydroEnergy

# Example usage to demonstrate inheritance and unique attributes
solar_farm = SolarEnergy("Desert Solar Farm", 5000, 0.22) # 5000 kW, 22% efficiency
wind_park = WindEnergy("Offshore Wind Park", 7500, 15) # 7500 kW, 15 m/s wind speed
hydro_plant = HydroEnergy("River Hydro Plant", 10000, 500) # 10000 kW, 500 m3/s water flow

print(f"\n{solar_farm._name}: Current Output = {solar_farm.get_current_output()} kW, Panel Efficiency = {solar_farm._panel_efficiency * 100}%")
print(f"{wind_park._name}: Current Output = {wind_park.get_current_output()} kW, Wind Speed = {wind_park._wind_speed} m/s")
print(f"{hydro_plant._name}: Current Output = {hydro_plant.get_current_output()} kW, Water Flow Rate = {hydro_plant._water_flow_rate} m3/s")
```

## Implement Polymorphic Energy Availability

### Subtask:
Define a polymorphic method `calculate_availability` in each subclass. This method will return the current available power, potentially with different logic for each energy source (e.g., solar dependent on sunlight, wind on wind speed, hydro on water flow). This will demonstrate polymorphism.


**Reasoning**:
The subtask requires defining a polymorphic `calculate_availability` method in the base class and overriding it in the subclasses with specific logic, then demonstrating its behavior. This code block directly implements all these requirements by modifying the existing classes and adding new examples to showcase the polymorphism.



```python
class EnergySource:
  def __init__(self, name, current_output):
    self._name = name  # Protected attribute for source name
    self._current_output = current_output  # Protected attribute for current power output

  def get_current_output(self):
    """Returns the current power output of the energy source."""
    return self._current_output

  def calculate_availability(self):
    """Abstract method to be overridden by subclasses to calculate available power."""
    raise NotImplementedError("Subclasses must implement calculate_availability method")

class SolarEnergy(EnergySource):
  def __init__(self, name, current_output, panel_efficiency):
    super().__init__(name, current_output)
    self._panel_efficiency = panel_efficiency # Unique attribute for SolarEnergy

  def calculate_availability(self):
    if self._panel_efficiency > 0.20:
      return self._current_output
    else:
      return self._current_output * 0.50

class WindEnergy(EnergySource):
  def __init__(self, name, current_output, wind_speed):
    super().__init__(name, current_output)
    self._wind_speed = wind_speed # Unique attribute for WindEnergy

  def calculate_availability(self):
    if self._wind_speed >= 10:
      return self._current_output
    else:
      return self._current_output * 0.70

class HydroEnergy(EnergySource):
  def __init__(self, name, current_output, water_flow_rate):
    super().__init__(name, current_output)
    self._water_flow_rate = water_flow_rate # Unique attribute for HydroEnergy

  def calculate_availability(self):
    if self._water_flow_rate >= 400:
      return self._current_output
    else:
      return self._current_output * 0.90

# --- Demonstrating Polymorphism ---
print("\n--- Demonstrating Polymorphic Availability ---")

# Solar Energy examples
solar_high_efficiency = SolarEnergy("High-Efficiency Solar Farm", 6000, 0.25) # Efficiency > 0.20
solar_low_efficiency = SolarEnergy("Low-Efficiency Solar Farm", 4000, 0.18)   # Efficiency <= 0.20

print(f"{solar_high_efficiency._name}: Current Output = {solar_high_efficiency.get_current_output()} kW, Efficiency = {solar_high_efficiency._panel_efficiency*100}%, Available Power = {solar_high_efficiency.calculate_availability():.2f} kW")
print(f"{solar_low_efficiency._name}: Current Output = {solar_low_efficiency.get_current_output()} kW, Efficiency = {solar_low_efficiency._panel_efficiency*100}%, Available Power = {solar_low_efficiency.calculate_availability():.2f} kW")

# Wind Energy examples
wind_high_speed = WindEnergy("High-Wind Park", 8000, 12) # Wind speed >= 10 m/s
wind_low_speed = WindEnergy("Low-Wind Park", 5000, 8)   # Wind speed < 10 m/s

print(f"{wind_high_speed._name}: Current Output = {wind_high_speed.get_current_output()} kW, Wind Speed = {wind_high_speed._wind_speed} m/s, Available Power = {wind_high_speed.calculate_availability():.2f} kW")
print(f"{wind_low_speed._name}: Current Output = {wind_low_speed.get_current_output()} kW, Wind Speed = {wind_low_speed._wind_speed} m/s, Available Power = {wind_low_speed.calculate_availability():.2f} kW")

# Hydro Energy examples
hydro_high_flow = HydroEnergy("High-Flow Hydro Plant", 12000, 450) # Water flow >= 400 m3/s
hydro_low_flow = HydroEnergy("Low-Flow Hydro Plant", 9000, 350)   # Water flow < 400 m3/s

print(f"{hydro_high_flow._name}: Current Output = {hydro_high_flow.get_current_output()} kW, Water Flow Rate = {hydro_high_flow._water_flow_rate} m3/s, Available Power = {hydro_high_flow.calculate_availability():.2f} kW")
print(f"{hydro_low_flow._name}: Current Output = {hydro_low_flow.get_current_output()} kW, Water Flow Rate = {hydro_low_flow._water_flow_rate} m3/s, Available Power = {hydro_low_flow.calculate_availability():.2f} kW")
```

## Implement Energy Allocation Logic

### Subtask:
Develop a function or class method that takes a list of energy sources and a demand. It will use conditional logic to prioritize energy sources (e.g., prioritize hydro, then solar, then wind) and allocate power using loops until the demand is met or all sources are depleted.


**Reasoning**:
The subtask requires implementing the `allocate_energy` function as described in the instructions, which includes categorizing energy sources, prioritizing them, and allocating power until demand is met or sources are depleted. This code block defines the function and provides example usage to demonstrate its functionality.



```python
def allocate_energy(energy_sources, demand):
  allocated_power = {}
  unmet_demand = demand

  hydro_sources = []
  solar_sources = []
  wind_sources = []

  # Categorize energy sources
  for source in energy_sources:
    if isinstance(source, HydroEnergy):
      hydro_sources.append(source)
    elif isinstance(source, SolarEnergy):
      solar_sources.append(source)
    elif isinstance(source, WindEnergy):
      wind_sources.append(source)

  # Define prioritization order
  prioritized_categories = [hydro_sources, solar_sources, wind_sources]

  # Allocate power based on prioritization
  for category in prioritized_categories:
    if unmet_demand <= 0:
      break # Demand met
    for source in category:
      if unmet_demand <= 0:
        break # Demand met

      available_power = source.calculate_availability()
      power_to_allocate = min(unmet_demand, available_power)

      allocated_power[source._name] = power_to_allocate
      unmet_demand -= power_to_allocate

  return allocated_power, unmet_demand

# --- Example Usage ---
print("\n--- Demonstrating Energy Allocation Logic ---")

# Instantiate some energy sources
sources_for_allocation = [
    SolarEnergy("Small Solar Farm", 1000, 0.20),       # 1000 * 0.5 = 500 kW available
    WindEnergy("Coastal Wind Farm", 2000, 15),       # 2000 kW available
    HydroEnergy("Mountain Hydro Plant", 3000, 300),   # 3000 * 0.9 = 2700 kW available
    SolarEnergy("Large Solar Farm", 4000, 0.25),       # 4000 kW available
    WindEnergy("Inland Wind Farm", 1500, 8),         # 1500 * 0.7 = 1050 kW available
    HydroEnergy("Valley Hydro Plant", 2500, 450)    # 2500 kW available
]

demand1 = 7000 # kW
print(f"\nAttempting to meet demand of {demand1} kW")
allocated1, unmet1 = allocate_energy(sources_for_allocation, demand1)
print("Allocated Power:", allocated1)
print(f"Unmet Demand: {unmet1} kW")

demand2 = 3000 # kW
print(f"\nAttempting to meet demand of {demand2} kW")
allocated2, unmet2 = allocate_energy(sources_for_allocation, demand2)
print("Allocated Power:", allocated2)
print(f"Unmet Demand: {unmet2} kW")

demand3 = 15000 # kW
print(f"\nAttempting to meet demand of {demand3} kW")
allocated3, unmet3 = allocate_energy(sources_for_allocation, demand3)
print("Allocated Power:", allocated3)
print(f"Unmet Demand: {unmet3} kW")
```

## Simulate and Report

### Subtask:
Simulate the energy allocation for a given demand, printing which sources are used and how much power is drawn from each, along with any unmet demand.


**Reasoning**:
The subtask requires defining a new function `generate_allocation_report` to present the results of the energy allocation simulation, and then calling it with the specified variables. This code block will define and execute the function.



```python
def generate_allocation_report(original_demand, allocated_power, unmet_demand):
  print(f"\n--- Energy Allocation Report for Demand: {original_demand} kW ---")

  if not allocated_power:
      print("No power was allocated from any source.")
  else:
      print("Power Allocated From:")
      for source_name, power in allocated_power.items():
          print(f"  - {source_name}: {power:.2f} kW")

  total_allocated = sum(allocated_power.values())
  print(f"\nTotal Power Allocated: {total_allocated:.2f} kW")
  print(f"Unmet Demand: {unmet_demand:.2f} kW")

  if unmet_demand <= 0:
      print("Conclusion: The energy demand was successfully met.")
  else:
      print("Conclusion: The energy demand was not fully met, and additional resources are required.")
  print("---------------------------------------------------")

# Call the function for demand3, allocated3, and unmet3
generate_allocation_report(demand3, allocated3, unmet3)
```
