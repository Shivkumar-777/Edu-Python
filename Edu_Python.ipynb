```python
weekly_temp = [25, 27, 28, 26, 24, 24, 30, 29]

print(f"Weekly temperatures : {weekly_temp}")

weekly_temp.remove(25)
print(f"Update 1 : Weekly temperatures : {weekly_temp}")

weekly_temp.insert(0, 23)
print(f"Update 2 : Weekly temperatures : {weekly_temp}")

weekly_temp.append(25)
print(f"Update 3 : Weekly temperatures : {weekly_temp}")

tuple_wt = tuple(weekly_temp)
tuple_wt.insert(0, 23)
print(f"Update 4 : Weekly temperatures : {tuple_wt}")
```

```python
# Conditional Statements

num1 = int(input("Enter first number : "))
num2 = int(input("Enter second number : "))

if num1 > num2:
    print(f"{num1} is greater than {num2}")

elif num1 < num2:
    print(f"{num1} is less than {num2}")

else:
    print(f"{num1} is equal to {num2}")
```

```python
# Calculator

num1 = int(input("Enter first number : "))
num2 = int(input("Enter second number : "))
operator = input("Enter operator : ")

if operator == '+':
    print(f"{num1} + {num2} = {num1 + num2}")

elif operator == '-':
    print(f"{num1} - {num2} = {num1 - num2}")

elif operator == '*':
    print(f"{num1} * {num2} = {num1 * num2}")

elif operator == '/':
    print(f"{num1} / {num2} = {num1 / num2}")

else:
    print("Invalid operator")
```
